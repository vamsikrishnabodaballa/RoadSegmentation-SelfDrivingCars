```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
from keras.models import load_model
%matplotlib notebook
```

```python
model = load_model('C:/Users/DELL/jupyter notebooks/road segmentation/final-road-seg-model-v2.h5')
```

```python
image = imageio.imread('C:/Users/DELL/jupyter notebooks/road segmentation/CameraRGB/F61-76.png')
seg =imageio.imread('C:/Users/DELL/jupyter notebooks/road segmentation/CameraSeg/F61-76.png')
seg = np.array([max(seg[i, j]) for i in range(seg.shape[0]) for j in range(seg.shape[1])]).reshape(image.shape[0], image.shape[1])
my_preds = model.predict(np.expand_dims(image, 0))
my_preds = my_preds.flatten()
my_preds = np.array([1 if i >= 0.5 else 0 for i in my_preds])
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].imshow(my_preds.reshape(600, 800))
ax[0].set_title('Prediction')
ax[1].imshow(image.reshape(600, 800,3))
ax[1].set_title('Image')
```
